```python
import pandas as pd
install_df = pd.read_excel(
    "C:/Users/Udith P Kidiyoor/Downloads/Fresher_MBA/Fresher_MBA/Test_Data_2026_Fresher.xlsx",
    sheet_name="SETA_InstallSource"
)
txn_df = pd.read_excel(
    "C:/Users/Udith P Kidiyoor/Downloads/Fresher_MBA/Fresher_MBA/Test_Data_2026_Fresher.xlsx",
    sheet_name="SETA_TransactionData"
)
```

```python
install_df = install_df.rename(columns={
    "UserID": "user_id",
    "Install Source": "install_source"
})

txn_df = txn_df.rename(columns={
    "UserID": "user_id",
    "Transaction Value (USD)": "txn_value"
})
```

```python
install_df = install_df.drop_duplicates(subset="user_id")
```

```python
df = txn_df.merge(
    install_df,
    on="user_id",
    how="left"
)
```

```python
install_kpis = (
    df.groupby("install_source")
      .agg(
          total_revenue=("txn_value", "sum"),
          total_users=("user_id", "nunique")
      )
      .reset_index()
)

install_kpis["ARPU"] = (
    install_kpis["total_revenue"] /
    install_kpis["total_users"]
)

install_kpis
```

```python
top_3 = install_kpis.sort_values("ARPU", ascending=False).head(3)
bottom_3 = install_kpis.sort_values("ARPU", ascending=True).head(3)

top_3
```
